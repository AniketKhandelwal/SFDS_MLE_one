```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#to generate random number using numpy use np.random module
experience = np.random.randint(low = 2, high = 11, size = 500) #sampling of uniform distribution
```

```python
plt.hist(experience)
```

```python
salary = 2*experience + 3 + np.random.normal(loc=0, scale=2, size=500)
```

```python
plt.scatter(experience, salary) 
plt.xlabel("Number of years of experience (X)")
plt.ylabel("Salary in Lakhs of Rupees (Y)")
```
